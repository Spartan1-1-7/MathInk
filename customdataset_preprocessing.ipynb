```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('train_final.csv', header=None)  # if there's no header

# Rename the last column to 'label' and others to 'pixel0'...'pixel783'
df.columns = [f'pixel{i}' for i in range(784)] + ['label']

# Check for missing values
missing_values = df.isnull().sum().sum()
if missing_values > 0:
    df = df.dropna()  # or use df.fillna(method='ffill') if that makes more sense for your case

# Ensure pixel values are numeric and scaled (optional but useful)
df.iloc[:, :-1] = df.iloc[:, :-1].astype(np.float32) / 255.0

# Ensure labels are integers
df['label'] = df['label'].astype(int)

# Optionally: check for duplicates
df = df.drop_duplicates()

# Save cleaned version
df.to_csv('train_cleaned.csv', index=False)

print("Dataset cleaned and saved as 'train_cleaned.csv'")
```

```python
import pandas as pd

# Load the CSV file
file_path = '/mnt/data/train_final.csv'
df = pd.read_csv('train_final.csv')

# Display first few rows to understand the structure
print("First 5 rows of the dataset:")
print(df.head())

# Check for missing values by column
print("\nMissing values by column:")
missing_by_col = df.isnull().sum()
print(missing_by_col[missing_by_col > 0])

# Check for missing values by row
print("\nMissing values by row:")
missing_by_row = df.isnull().sum(axis=1)
rows_with_missing = df[missing_by_row > 0]
print(rows_with_missing)

# Optional: Show indices of rows with NaNs
print("\nIndices of rows with missing values:")
print(rows_with_missing.index.tolist())
```
